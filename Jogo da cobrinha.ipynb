```python
import pygame
from pygame.locals import *
from sys import exit
from random import randint

pygame.init()

pygame.display.set_caption('Jogo da cobrinha')

lar = 640
alt = 480

#lar_cobra = 20
#alt_cobra = 50

tela = pygame.display.set_mode((lar,alt))

x = lar/2 - 20/2
y = alt/2 - 50/2

velocidade = 10
x_controle = velocidade
y_controle = 0

x_presa = randint(40,600)
y_presa = randint(50,430)

relogio = pygame.time.Clock()

fonte = pygame.font.SysFont('arial',35,True,True)
pontos = 0

pygame.mixer.music.set_volume(0.07)
musica = pygame.mixer.music.load('BoxCat Games - Mt Fox Shop.mp3')
pygame.mixer.music.play(-1)

som_colisao = pygame.mixer.Sound('smw_coin.wav')
som_morte = pygame.mixer.Sound('smb_gameover.wav')

lista_cobra = []
lar_cobra = 5

morre = False
def renicia():
    global pontos,lar_cobra,x,y,lista_cobra,lista_c,x_presa,y_presa,morre
    pontos = 0
    lar_cobra = 5
    x = lar/2 - 20/2
    y = alt/2 - 50/2
    lista_cobra = []
    lista_c = []
    x_presa = randint(40,600)
    y_presa = randint(50,430)
    morre = False
    

def aumenta_cobra(lista_cobra):
    for xey in lista_cobra:
        pygame.draw.rect(tela,(0,255,0),(xey[0],xey[1],20,20))

while True:
    
    relogio.tick(30)
    tela.fill((0,0,0))
    
    mensagem = f'Pontos: {pontos}'
    texto = fonte.render(mensagem,True,(255,255,255))
    
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            exit()
        if event.type == KEYDOWN:
            if event.key == K_a:
                if x_controle == velocidade:
                    pass
                else:
                    x_controle = -velocidade
                    y_controle = 0
            if event.key == K_d:
                if x_controle == -velocidade:
                    pass
                else:
                    x_controle = velocidade
                    y_controle = 0
            if event.key == K_w:
                if y_controle == velocidade:
                    pass
                else:
                    x_controle = 0
                    y_controle = -velocidade
            if event.key == K_s:
                if y_controle == -velocidade:
                    pass
                else:
                    x_controle = 0
                    y_controle = velocidade
    x = x + x_controle
    y = y + y_controle
    
    cobra = pygame.draw.rect(tela,(0,255,0),(x,y,20,20))
    presa = pygame.draw.rect(tela,(255,0,0),(x_presa,y_presa,20,20))
    
    if cobra.colliderect(presa):
        x_presa = randint(40,600)
        y_presa = randint(50,430)
        pontos += 1
        som_colisao.play()
        lar_cobra += 1
        
    lista_c = []
    lista_c.append(x)
    lista_c.append(y)
    
    lista_cobra.append(lista_c)
    
    if lista_cobra.count(lista_c) > 1:
        fonte2 = pygame.font.SysFont('arial',25,True,True)
        mensagem2 = 'Morreu, aperte a tecla R para reniciar'
        texto2 = fonte2.render(mensagem2,True,(0,0,0))
        ret_texto = texto2.get_rect()
        pygame.mixer.music.pause()
        som_morte.play()
        morre = True
        while morre:
            tela.fill((255,255,255))
            for event in pygame.event.get():
                if event.type == QUIT:
                    pygame.quit()
                    exit()
                if event.type == KEYDOWN:
                    if event.key == K_r:
                        renicia()
                        pygame.mixer.music.play()
            
            ret_texto.center = (lar//2,alt//2)
            tela.blit(texto2,ret_texto)
            pygame.display.update()
    
    if x < 0:
        x = lar
    if x > lar:
        x = 0
    if y < 0:
        y = alt
    if y > alt:
        y = 0
    
    if len(lista_cobra) > lar_cobra:
        del lista_cobra[0]
    
    aumenta_cobra(lista_cobra)
    
    tela.blit(texto,(475,40))
    pygame.display.update()
```
